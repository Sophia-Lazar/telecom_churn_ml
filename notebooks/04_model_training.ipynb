```python
import numpy as np 
import pandas as pd 
import joblib
```

### Load processed data

```python
x_train,x_test,y_train,y_test=joblib.load("..\data\processed\split_data.pkl")
```

### Feature Scaling

Feature scaling is performed using StandardScalar for the Logistic Regression model.

Logistic regression is sensitive to the magnitude of input feature, and the scaling ensures all the features contribute equally to the learning process.

Tree based models such as RandomForest do not require scaling because they split data based on feature thresholds rather than destance calculations. Therefore scaling was applied only to the Logistic Regression Model.

The scalar was fitted on to the traing dataset and applied on to the test dataset to prevent data Leakage 

```python
from sklearn.preprocessing import StandardScaler

scalar=StandardScaler()

x_train_scaled=scalar.fit_transform(x_train)
x_test_scaled=scalar.transform(x_test)
```

### Logistic Regression with Class Weights

The dataset exibits class imbalance, where the number of non-churn customers is significantly higher than churn customers.

To address this imbalance the logistoc regression is trained using class_weight="balanced" parameter. this approach automatically adjust the class importance during training, giving higher weight to minority class(churn) so that the model pays more atention to it.

This method helps improve the models ability to detect churners without modifying the data.


```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

lr_balanced_pipeline=Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(class_weight="balanced",max_iter=2000))
])


lr_balanced_pipeline.fit(x_train,y_train)

joblib.dump(lr_balanced_pipeline,"../models/logistic_cw_pipeline.pkl")
```

### Logical regression with SMOTE

SMOTE (Synthetic Minority Oversampling Technique) was applied to generate synthetic samples for the minority class. This balances the dataset and helps the model to learn better patterns for predicting the churn customers.

```python
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline

lr_smote_pipeline=Pipeline([
    ("scaler",StandardScaler()),
    ("smote",SMOTE(random_state=42)),
    ("model",LogisticRegression(max_iter=1000))
])

lr_smote_pipeline.fit(x_train,y_train)
joblib.dump(lr_smote_pipeline,"../models/logistic_smote_pipeline.pkl")
```
