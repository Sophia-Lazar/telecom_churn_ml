```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("..\data\processed\cleaned_data.csv")
df.head(5)
```

### Removing Unnecessary columns

```python
df=df.drop("customerID",axis=1)
```

The customerID was dropped because it is unique identifier and does not contain predictive information for the "Churn model"

### Encode categorical variables

```python
df=pd.get_dummies(df,drop_first=True)
```

Categorical features were converted into numerial format to allow machine learning models to learn them

### Split data

```python
from sklearn.model_selection import train_test_split

x=df.drop("Churn",axis=1)
y=df['Churn']

x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2,stratify=y)
```

The dataset is divided into train and test sets using an 80/20 split.
stratified sampling is used to maintain the class distribution of the target variable.

save the data

```python
import joblib

joblib.dump((x_train,x_test,y_train,y_test),"..\data\processed\split_data.pkl")
```
