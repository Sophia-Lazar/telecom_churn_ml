### Spliting train and test data

```python
from  sklearn.model_selection import train_test_split

x=df.drop(df["Churn"],axis=1)
y=df["Churn"]

x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2,stratify=y)
```

```python
from sklearn.preprocessing import StandardScaler

scalar=StandardScaler()

x_train_scaled=scalar.fit_transform(x_train)
x_test_scaled=scalar.transform(x_test)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

model=LogisticRegression(max_iter=1000)
model.fit(x_train_scaled,y_train)
```

```python
y_pred=model.predict(x_test_scaled)
```

Evaluate Model

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

print("Accuracy",accuracy_score(y_test,y_pred))
print("confusion matrix",confusion_matrix(y_test,y_pred))
print("classification report",classification_report(y_test,y_pred))
```

### Model Performance summary

The model achieves 80% overall accuracy.

For churn prediction (class 1):
The recall is 57%, which means that the model identifies more than half of the customers who will churn
The precision is 65%,indicating 65% of the flagged customers are truly churn.

While the performance is reasonable, the model misses 43% of the churn.
Improving the recall allows the business to identify more customers who are at the risk of churning.



#### Handling class imbalance

since churn class(1) is smaller. Using class_weight method

```python
from sklearn.linear_model import LogisticRegression

model_cw=LogisticRegression(class_weight="balanced", max_iter=2000)
model_cw.fit(x_train_scaled,y_train)
```

```python
y_pred_cw=model_cw.predict(x_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

print("accuracy_score:",accuracy_score(y_test,y_pred_cw))
print("confusion_matrix:",confusion_matrix(y_test,y_pred_cw))
print("classification_report:",classification_report(y_test,y_pred_cw))
```

Accuracy decreased to 72%, but the recall for class(1) increased to 80%.

Using SMOTE

```python
from imblearn.over_sampling import SMOTE

sm=SMOTE(random_state=42)

x_res,y_res=sm.fit_resample(x_train_scaled,y_train)
```

```python
print("before SMOTE",y_train.value_counts())
print("after SMOTE",y_res.value_counts())
```

```python
from sklearn.linear_model import LogisticRegression

model=LogisticRegression(max_iter=1000)

model.fit(x_res,y_res)
```

```python
y_pred=model.predict(x_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score

print("confusion_matrix",confusion_matrix(y_test,y_pred))
print("Accuracy score",accuracy_score(y_test,y_pred))
print("classification_report",classification_report(y_test,y_pred))
```

#### Comparison

Class_weight: accuracy- 72%, recall(class 1)-80%, f1_score- 61%
Using SMOTE: accuracy- 73%, recall(class 1)-78%, f1_score- 61%

Even though the accuracy increased by 1% the recall decreased by 2% hence using clas_weight is better than using SMOTE

## Random Forest

#### Logistic Regression
- Linear model
- Struggles with complex relationships

#### Randon forest
- Captures non-linear patterns
- Handles categorical encoded features well
- works well for imbalance with weight balance
- usually improves recall + f1

```python
from sklearn.ensemble import RandomForestClassifier

rf= RandomForestClassifier(n_estimators=200,class_weight="balanced",random_state=42)

rf.fit(x_train,y_train)

y_pred_rf=rf.predict(x_test)
```

##### Model Performance

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix

print("accuracy_score: ",accuracy_score(y_test,y_pred_rf))
print("classification_report: ",classification_report(y_test,y_pred_rf))
print("confusion_matrix: ",confusion_matrix(y_test,y_pred_rf))
```

#### Conclusion

Logistic regression with Class_weight="balanced": 
accuracy- 72%, recall(class 1)-80%, f1_score- 61%

Random forest with calss_weight="balanced":
accuracy- 79%, recall(class 1)-50%, f1_score- 56%

accuracy is increased to 79% but recall and f1_score decreased so much 

since recall is so important in churn problem we cannot use this model still logistic regression with class weight is better

### To make random forest better apply hyper parameter tuning


```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# define parameter grid

param_grid={
    'n_estimators':[100,200,300],
    'max_depth': [None,10,20,30],
    'min_samples_split':[2,5,10],
    'min_samples_leaf':[1,2,4],
    'class_weight': ["balanced"]
}
```

```python
rf=RandomForestClassifier(random_state=42)

grind_search= GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=3,
    scoring='recall',
    n_jobs=1
)

grind_search.fit(x_train,y_train)
```

```python
best_rf=grind_search.best_estimator_

y_pred_brf=best_rf.predict(x_test)

from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

print("accuracy_score",accuracy_score(y_test,y_pred_brf))
print("classification_report",classification_report(y_test,y_pred_brf))
print("confusion_matrix",confusion_matrix(y_test,y_pred_brf))
```

So far Random forest(Tuned) is the best model

#### Feature importance

Model checks how much each feature contributes to reducing prediction error in the trees

```python
import pandas as pd

feature_importance=pd.Series(best_rf.feature_importances_,index=x_train.columns)
feature_importance=feature_importance.sort_values(ascending=False)

print(feature_importance)
```

#### ROC-AUC

```python
from sklearn.metrics import roc_curve,roc_auc_score
import matplotlib.pyplot as plt

y_prob = best_rf.predict_proba(x_test)[:,1]

fpr, tpr, thresholds = roc_curve(y_test,y_prob)

auc_score =roc_auc_score(y_test,y_prob)
print("auc score: ",auc_score)
```

```python
plt.plot(fpr,tpr,label= "Random forest (AUC = %.2f)" % auc_score)
plt.plot([0,1],[0,1],'--')
plt.xlabel("False positive rate")
plt.ylabel("True positive rate")
plt.title("ROC curve")
plt.legend()
plt.show()
```
